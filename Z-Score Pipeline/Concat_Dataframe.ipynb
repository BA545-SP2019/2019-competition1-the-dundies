```python
# Importing packages and importing csv files to respective variables

import pandas as pd
data = pd.read_csv("ReadyDF_NORMAL", index_col=0)
Targets = pd.read_csv("ReadyDF_TARGETS", index_col=0)
```

```python
# We will be dropping everything but columns 'Y1' and 'Y2'
Targets = Targets.drop(['I1','I2','I3','P_IPO','P_L','P_1Day','P_H'],axis=1)
Targets.head(2)
```

```python
# Concatinating the dataframes into one dataframe
DF_Use = pd.concat([data,Targets], axis=1)
DF_Use.head(2)
```
